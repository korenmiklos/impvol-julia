```julia
p = load("../experiments/CES2/common_parameters.jld2")["parameters"]
data = load("../data/impvol_data.jld2")
```

```julia
function relative(series)
    return series .- series[:,:,:,1:1]
end
function check(p, data, country=21, sector=24)
    P_US = ImpvolEquilibrium.CES_price_index(p[:nu_njt], p[:p_sectoral][1:1,end:end,:,:], p[:sigma])
    FES = p[:nu_njt] .* (p[:p_sectoral] ./ (data["pwt"] .* P_US)) .^ (1-p[:sigma])
    ratio = log.(p[:nu_njt] ./ FES) ./ (p[:sigma]-1)
    series = (relative(ratio), relative(log.(p[:p_sectoral])), relative(log.(p[:A])))
    X = zeros(p[:T], length(series))
    for s = 1:length(series)
        X[:,s] .= series[s][1,country,sector,:]
    end
    plot(X, label=["ratio", "P", "A"])
end
```

```julia
check(p, data, 25, 24)
```
