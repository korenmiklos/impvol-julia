```julia
using FileIO, CSV, DataFrames
```

```julia
function get_productivity(experiment)
        p = load("../experiments/$experiment/common_parameters.jld2")["parameters"]
        return p[:A]
end
```

```julia
A05 = get_productivity("CES05")
A20 = get_productivity("CES2")
A10 = get_productivity("baseline")
```

```julia
A10 ./ A05
```

```julia
function save_data(data, name)
    _, N, J, T = size(data)
    for n=1:N
        CSV.write("$name$n.csv", DataFrame(data[1,n,:,:]'))
    end
end
```

```julia
save_data(A05, "A05")
```

```julia
save_data(A10, "A10")
```

```julia
save_data(A20, "A20")
```
