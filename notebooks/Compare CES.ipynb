```julia
function comparative_advantage(r)
    A = r[:A_njs]
    P = r[:P_njs]
    share = r[:real_GDP] ./ sum(r[:real_GDP], 3)
    return sum(share .* log.(P), 3)[1,:,1,1]
end
```

```julia
plot([comparative_advantage(r05) comparative_advantage(r1) comparative_advantage(r2)])
```
